## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf # I am using Yahoo Finance
# 'pip install yfinance' if library is not installed.
```

## Data Extraction

```python
# Function definition for extracting data
def extracting_data(symbol,start_date,end_date):
    try:
        # Using yfinance to extract historical data for the symbol and date range specified
        stock_data = yf.download(symbol + ".NS", start=start_date, end=end_date) # Using '.NS' i.e NSE(National Stock Exchange)
        return stock_data
    
    # Handles any exceptions if occurred during data extraction
    except Exception as e:
        print("Error occurred during data extraction is :", str(e))
        return None
```

```python
# Defining the symbol,start date and end date for data extraction
symbol = 'Reliance'  # I am extracting stock details of Reliance company here
start_date = '2022-01-23'  # from 23rd of January, 2022
end_date = '2023-06-23'  # to 23rd of June, 2023 (format: YYYY-MM-DD)
```

```python
# Extracting data by calling function defined above i.e.'extracting_data' function
stock_data = extracting_data(symbol,start_date,end_date)
print(stock_data.head(5))
```

```python
df=pd.DataFrame(stock_data)
df
```

## Calculating Average Volume

```python
# Function defintion for calculating the average daily trading volume
def average_volume(stock_data):
    try:
        # Calculating the average daily trading volume using pandas
        average_volume = stock_data['Volume'].mean()
        return average_volume
    
    # Handles any exceptions if occurred during calculating
    except Exception as e:
        print("Error occurred during Data Processing is : ", str(e))
        return None
```

```python
# Function call
print("Average Daily trading volume for the selected period is:",average_volume(df))
```

## Data Visualization

```python
width = 3 # width of real body
width2 = 0.1  # width of shadow

fig, ax = plt.subplots(figsize=(60,30))

# finding the rows that are bullish
dfup = df[df.Close >= df.Open]

# finding the rows that are bearish
dfdown = df[df.Close < df.Open]

# plotting the bullish candle stick
ax.bar(dfup.index, dfup.Close - dfup.Open, width, bottom = dfup.Open, edgecolor='green', color='green'
       ,label='Bullish (prices ready to move high)')
ax.bar(dfup.index, dfup.High - dfup.Close, width2, bottom = dfup.Close, edgecolor='g', color='green')
ax.bar(dfup.index, dfup.Low - dfup.Open, width2, bottom = dfup.Open, edgecolor='g', color='green')

# plotting the bearish candle stick
ax.bar(dfdown.index, dfdown.Close - dfdown.Open, width, bottom = dfdown.Open, edgecolor='r', color='red'
       ,label='Bearish (prices ready to move low)')
ax.bar(dfdown.index, dfdown.High - dfdown.Open, width2, bottom = dfdown.Open, edgecolor='r', color='red')
ax.bar(dfdown.index, dfdown.Low - dfdown.Close, width2, bottom = dfdown.Close, edgecolor='r', color='red')

# Graph customization
plt.xticks(fontsize=40,rotation=45)
plt.yticks(fontsize=40)
plt.legend(fontsize=40)
plt.xlabel('Date',fontsize=45,fontweight ="bold")
plt.ylabel('Price',fontsize=45,fontweight ="bold")
plt.title('Graph of Price vs Date in Selected date range',fontsize=50)
plt.grid(color='black',linestyle='--',linewidth=2)

plt.show()
```
